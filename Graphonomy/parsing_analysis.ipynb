```python
from PIL import Image
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
imagePath = '/data/muxiangyu/pythonPrograms/Graphonomy/data/datasets/CIHP_4w/Category_rev_ids/0000002.png'

img = Image.open(imagePath)
print(img.mode)
color_np = np.array([[0,0,0]
                , [128,0,0], [255,0,0], [0,85,0], [170,0,51], [255,85,0], [0,0,85], [0,119,221], [85,85,0], [0,85,85], [85,51,0], [52,86,128], [0,128,0]
                , [0,0,255], [51,170,221], [0,255,255], [85,255,170], [170,255,85], [255,255,0], [255,170,0]])
bin_colormap = np.random.randint(0, 255, (256, 3))
print(color_np)
print(bin_colormap)
bin_colormap = bin_colormap.astype(np.uint8)
img = img.convert('P')
img.putpalette(bin_colormap)
plt.imshow(img)
plt.show()
img_np = np.array(img)
print(img_np.shape)
print(img_np.dtype)
print(img_np.min(), img_np.max())

pixelset = set()
for i in range(img_np.shape[0]):
    for j in range(img_np.shape[1]):
        pixelset.add(img_np[i][j])
print(pixelset)

img_np = (img_np==5).astype(np.uint8)
plt.imshow(img_np)
plt.show()
```
